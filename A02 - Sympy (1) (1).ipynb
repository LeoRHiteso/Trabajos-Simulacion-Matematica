$f(x)=6-2x+\sqrt{18-2x^2}$, para $-3\leq x\leq 3$

```python
import sympy as sp
import numpy as np
x= sp.var("x")
def f(x):
    return 6-2*x + (18-(2*x**2))**(1/2)
dy= sp.diff(f(x),x)
d2y=sp.diff(dy,x)
pc = sp.solve(dy, x)
pc
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-3,3,0.1)
y = 6-2*x + (18-(2*x**2))**(1/2)
plt.plot(x,y)
plt.grid()
plt.plot()
```

El punto máximo se encuentra en -2.4495. Al ser integral, sabemos que por lo menos, por la izquierda no existe otro valor máximo ni mínimo. Este es un máximo global.

$f(x)=\frac{1}{4}x^4+\frac{1}{3}x^3-x^2$ en el intervalo $[-3,1.7]$

### El perímetro de un triángulo isósceles es de $10 cm$. ¿Cuánto deben medir sus lados para que el volumen del cuerpo generado por la rotación del triángulo en torno a su base sea el mayor posible? (Volumen de un cono $= \frac{1}{3}\pi r^2 h$, donde $r$ es el radio de la base y $h$ la altura del cono).

 **información dada**  
 perímetro = p = 10cm  
 **Información general**  
 fórmula del volumen =  $V = \frac{1}{3}\pi r^2 h$  
 r = radio de la base y h = altura del cono
 $L^2$  
 

```python
import sympy as sp
l = sp.var("l")
b = sp.var("b")
r = sp.var("r")
h = sp.var("h")
def V(r, h):
    return sp.pi*r**2*h/3
V(r, h)
```

```python
restriccion = b + 2*l - 10
radio =- b/2
pitagoras = l**2 - r**2 - h**2
```

```python
altura = sp.solve(pitagoras, h)[1]
altura
```

```python
V_lb= V(r,h).subs(h, altura).subs(r, radio)
```

```python
l_despeje = sp.solve(restriccion,l)[0]
l_despeje
```

```python
V(r, h).subs(h, altura).subs(r, radio).subs(l, l_despeje)
```

```python
V_b= V_lb.subs(l, l_despeje)
V_b
```

```python
dV = sp.diff(V_b,b)
dV
```

**Por lo tanto los lados deben medir**

```python
sp.solve(dV, b)[1]
```

### Disponemos de una barra de aluminio de 6 metros para construir una portería de fútbol. Si queremos que el área de la portería sea máxima, ¿cuánto deben medir los postes y el travesaño? ###

```python
x = sp.var("x")
y= sp.var("y")

def A(x,y):
    return x*y
restriccion = x + 2*y - 6
```

```python
x_y = sp.solve(restriccion, x)[0]
x_y
```

```python
A_y = A(x,y).subs(x, x_y)
A_y
```

```python
dA = sp.diff(A_y , y)
dA
```

```python
sp.solve(dA, y)[0]
```

Los postes deben medir 1.5m cada uno

```python
x_y.subs(y, 1.5)
```

**Respuesta:** Debe medir el travesaño $3m$

2.1 Defina una función, cuyos argumentos de entrada sean:


1. el tiempo $t$,
2. el arreglo $[x,\dot{x}]$.

Y la salida:

1. el arreglo $[\dot{x},\ddot{x}]$.

  Considerando el argumento de la frecuencia natural $\omega_0$ debe ser una variable global.

```python
from scipy.integrate import solve_ivp
def oscilador(t, X, w0):
    x = X[0]
    v = X[1]
    dx = v
    dv = -w0**2*x
    return [dx, dv]
tspan = [0,10]
x0 = np.array([0.5,0.1])
sol_os = solve_ivp(oscilador, tspan, x0 , args=(3,), max_step = 0.01)
sol_os
```

Grafique

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize = (7, 7))
plt.plot(sol_os.t, sol_os.y[1,:], lw = 3, label = "Velocidad")
plt.xlabel("Velocidad")
plt.ylabel("Tiempo")
plt.grid()
plt.plot(sol_os.t, sol_os.y[0,:], color="green" , lw=3, label = "Posición")
plt.legend(loc="best")
```
